### There are missing values such as None or NaN How many are there for each column?

```python
missingVal_cnt = nfl_data.isnull().sum()
```

```python
# the number of missing values from frst 20 columns
missingVal_cnt[0:20]
```

```python
# 25% of total NFL data point is null value

# nfl_data.shape returns (number of rows, number of columns)
# np.product(nfl_data.shape) returns number_rows*number_columns, which is the total_cells
totalCells = np.product(nfl_data.shape) 
totalMissing = missingVal_cnt.sum()

(float(totalMissing)/float(totalCells))*100
```

```python
nfl_data.shape[1] # num of rows
```

```python
nfl_data.shape[0] # num of col
```

```python
def missingValProportion(data):
    totalNullCnt = data.isnull().sum().sum()
    totalCellCnt = np.product(data.shape)
    
    proportion = (float(totalNullCnt)/float(totalCellCnt))*100
    
    return proportion
```

```python
# 26% of total SF data point is null value
missingValProportion(sf_permits)
```

### Drop Missing Values
- Remove any rows or columns that contain missing values
- Quick and dirty way - might lose meaningful data

```python
# using pandas to remove all ROWS that contain a missing value
nfl_data.dropna()

# … and we removed everything LOL
```

```python
# using pandas to remove all COLUMNS with at least one missing value
nullDroppedCol = nfl_data.dropna(axis = 1)
nullDroppedCol.head()
```

### After Doing This How Many Columns Did We Lose?

```python
print "original num of col:", nfl_data.shape[1]
print "num of col after dropping null:", nullDroppedCol.shape[1]
```

__Repeating same thing for sf_permit dataset__

```python
# remove all ROWS that contains null
sf_permits.dropna()

# … and we removed everything again LOL
```

```python
# remove all COLUMNS that contains null
nullDroppedCol_sf = sf_permits.dropna(axis = 1)
nullDroppedCol_sf.head()
```

```python
print "original num of col:", sf_permits.shape[1]
print "num of col after dropping null:", nullDroppedCol_sf.shape[1]
```

### Filling In Missing Values Automatically Instead of Deleting all The Columns "IMPUTATION"

```python
# get a small subset (only several columns) of dataset to print conveniently 
nflSubset = nfl_data.loc[:,'EPA':'Season'].head()
nflSubset
```

```python
# replace all NaN with 0

nflSubset.fillna(0)
```

```python
# replace NaN with a value that comes directly after in the same col
# then replace all other null with 0
# useful when datasets have some sort of order to them

nflSubset.fillna(method = 'bfill', axis=0).fillna(0)
```

__Do same thing for sf_permit dataset__

```python
# subset the data
sfSubset = sf_permits.loc[:, 'Existing Construction Type':'Site Permit'].head()
sfSubset
```

```python
sfSubset.fillna('null')
```

```python
sfSubset.fillna(method = 'bfill', axis=0).fillna('null')
```
